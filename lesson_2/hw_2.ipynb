```python
# Удаление выбросов из столбца 'Value'
filtered_df, removed_percentage = utl.remove_extreme_outliers(df, "Value")
```

```python
### Вычисление квартилей для Value
first_quartile = df.Value.quantile(0.25)
third_quartile = df.Value.quantile(0.75)

# Вычисление межквартильного размаха (IQR)
iqr = third_quartile - first_quartile

# Определение нижней и верхней границ для выбросов
# Стандартный метод (множитель 1.5)
lower_bound_1_5 = first_quartile - 1.5 * iqr
upper_bound_1_5 = third_quartile + 1.5 * iqr

# Метод с множителем 3
lower_bound_3 = first_quartile - 3 * iqr
upper_bound_3 = third_quartile + 3 * iqr

# Вывод результатов
print("Метод с множителем 1.5:")
print(f"Нижняя граница: {lower_bound_1_5}")
print(f"Верхняя граница: {upper_bound_1_5}")

print("\nМетод с множителем 3:")
print(f"Нижняя граница: {lower_bound_3}")
print(f"Верхняя граница: {upper_bound_3}")

print(f"Первый квартиль (Q1): {first_quartile}")
print(f"Третий квартиль (Q3): {third_quartile}")
print(f"Межквартильный размах (IQR): {iqr}")

# Поиск выбросов по обоим методам
outliers_1_5 = df[(df["Value"] < lower_bound_1_5) | (df["Value"] > upper_bound_1_5)]
outliers_3 = df[(df["Value"] < lower_bound_3) | (df["Value"] > upper_bound_3)]

# print("\nВыбросы по методу 1.5:")
# print(outliers_1_5)

print("\nВыбросы по методу 3:")
outliers_3
```

### Напишите функцию, которая добавит колонку с разбиением возраста по группам: до 20, от 20 до 30, от 30 до 36 и старше 36. Посчитайте количество футболистов в каждой категории.

```python
test_df = df_without_duplicates.copy()
```

```python
test_df["Age_group"] = pd.cut(test_df.Age, [0, 20, 30, 36, float("inf")])
```

```python
age_group_stats = (
    test_df.groupby("Age_group", observed=True)["Age"]
    .agg(
        [
            "count",
        ]
    )
    .reset_index()
)
age_group_stats
```

```python
test_df = test_df.merge(age_group_stats, how="left", on=["Age_group"])
utl.show(test_df)
```
